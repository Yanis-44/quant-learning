```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv"
df = pd.read_csv(url)
df = df[["Date", "AAPL.Close"]]
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
```

```python
df["EMA_9"] = df["AAPL.Close"].ewm(span=9).mean()
df["EMA_21"] = df["AAPL.Close"].ewm(span=21).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["AAPL.Close"], label="Prix")
plt.plot(df["EMA_9"], label="EMA 9")
plt.plot(df["EMA_21"], label="EMA 21")
plt.legend()
plt.title("Stratégie EMA crossover (Apple)")
plt.grid(True)
plt.show()
```
